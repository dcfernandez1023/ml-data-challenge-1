```python
import pandas as pd
import matplotlib.pyplot as plt

train_df = pd.read_csv("train.csv")
train_df
```

## Columns I care about
- Activity Period
- GEO Summary
- GEO Region
- Activity Type Code
- Cargo Type Code
- Cargo Aircraft Type
- Cargo Metric TONS

```python
include = [
    'Activity Period',
    'GEO Summary',
#     'GEO Region',
    'Activity Type Code',
    'Cargo Type Code',
    'Cargo Aircraft Type',
    'Cargo Metric TONS'
]
exclude = [col for col in train_df.columns if col not in include]
train_df.drop(columns=exclude, inplace=True, axis=1)
train_df
```

## Converting 'Activity Period' to Dates

```python
train_df["Activity Period Year"] = train_df["Activity Period"].astype(str).str.slice(0, 4).astype(int)
train_df["Activity Period Month"] = train_df["Activity Period"].astype(str).str.slice(4, 6).astype(int)
train_df["Activity Period Day"] = 1

train_df["Activity Period Date"] = pd.to_datetime(pd.DataFrame({
    "year": train_df["Activity Period Year"],
    "month": train_df["Activity Period Month"],
    "day": train_df["Activity Period Day"]
}))
train_df
```

```python
train_df["Activity Period Date"].value_counts()
```

```python
# Sort by date
train_df.sort_values(by="Activity Period Date", inplace=True)
```

## EDA 

```python
year_periods = train_df["Activity Period Year"].value_counts().sort_index(axis=0)
year_periods.plot(kind="bar")
```

```python
geo_sum_counts = train_df["GEO Summary"].value_counts()
geo_sum_counts.plot(kind="bar")
```

```python
air_type_counts = train_df["Cargo Aircraft Type"].value_counts()
air_type_counts.plot(kind="bar")
```

```python
# plot some graphs :)
avg_vol_by_date = train_df.groupby("Activity Period Date")["Cargo Metric TONS"].agg("mean")
plt.xlabel("Time")
plt.ylabel("Avg. Cargo Metric TONS")
plt.title("Avg. Cargo Metric TONS Over Time")
avg_vol_by_date.plot()
plt.show()
```

```python
avg_carg_inter = train_df[train_df["GEO Summary"] == "International"].groupby("Activity Period Date")["Cargo Metric TONS"].agg("mean")
avg_carg_dom = train_df[train_df["GEO Summary"] == "Domestic"].groupby("Activity Period Date")["Cargo Metric TONS"].agg("mean")

avg_carg_inter.rename("International", inplace=True)
avg_carg_dom.rename("Domestic", inplace=True)
```

```python
avg_carg_inter.plot(legend=True)
avg_carg_dom.plot(legend=True)
plt.show()

# Seems like domestic cargo metric is greater than international on average
```

```python
avg_carg_enpl = train_df[train_df["Activity Type Code"] == "Enplaned"].groupby("Activity Period Date")["Cargo Metric TONS"].agg("mean")
avg_carg_depl = train_df[train_df["Activity Type Code"] == "Deplaned"].groupby("Activity Period Date")["Cargo Metric TONS"].agg("mean")

avg_carg_enpl.rename("Enplaned", inplace=True)
avg_carg_depl.rename("Deplaned", inplace=True)
```

```python
avg_carg_enpl.plot(legend=True)
avg_carg_depl.plot(legend=True)
plt.show()
```

```python
avg_carg_carg = train_df[train_df["Cargo Type Code"] == "Cargo"].groupby("Activity Period Date")["Cargo Metric TONS"].agg("mean")
avg_carg_mail = train_df[train_df["Cargo Type Code"] == "Mail"].groupby("Activity Period Date")["Cargo Metric TONS"].agg("mean")
avg_carg_expr = train_df[train_df["Cargo Type Code"] == "Express"].groupby("Activity Period Date")["Cargo Metric TONS"].agg("mean")

avg_carg_carg.rename("Cargo", inplace=True)
avg_carg_mail.rename("Mail", inplace=True)
avg_carg_expr.rename("Express", inplace=True)
```

```python
avg_carg_carg.plot(legend=True)
avg_carg_mail.plot(legend=True)
avg_carg_expr.plot(legend=True)
plt.show()

# Seems like the Cargo Type Code 'Cargo' has the highest average Cargo Metric TONS
```

```python
avg_carg_pass = train_df[train_df["Cargo Aircraft Type"] == "Passenger"].groupby("Activity Period Date")["Cargo Metric TONS"].agg("mean")
avg_carg_frei = train_df[train_df["Cargo Aircraft Type"] == "Freighter"].groupby("Activity Period Date")["Cargo Metric TONS"].agg("mean")
avg_carg_combi = train_df[train_df["Cargo Aircraft Type"] == "Combi"].groupby("Activity Period Date")["Cargo Metric TONS"].agg("mean")

avg_carg_pass.rename("Passenger", inplace=True)
avg_carg_frei.rename("Freighter", inplace=True)
avg_carg_combi.rename("Combi", inplace=True)
```

```python
avg_carg_pass.plot(legend=True)
avg_carg_frei.plot(legend=True)
avg_carg_combi.plot(legend=True)
plt.show()
```

# Conclusions from EDA
- For the `GEO Summary` field, the Domestic flights have a higher Cargo Metric TONS on avg
- For the `Cargo Type Code` the Cargo flights have a higher Cargo Metric TONS on avg, the Express flights were higher on avg until about 2013 and then the Mail flights were higher on avg afterwards
- For the `Cargo Aircraft Type` field, the Freighter flights were higher on avg throughout the years, and it looks like they stopped doing Combi flights after 2015

# Feature Engineering the Test Set

```python
test_df = pd.read_csv("test.csv")

# Activity Period Dates
test_df["Activity Period Year"] = test_df["Activity Period"].astype(str).str.slice(0, 4).astype(int)
test_df["Activity Period Month"] = test_df["Activity Period"].astype(str).str.slice(4, 6).astype(int)
test_df["Activity Period Day"] = 1

test_df["Activity Period Date"] = pd.to_datetime(pd.DataFrame({
    "year": test_df["Activity Period Year"],
    "month": test_df["Activity Period Month"],
    "day": test_df["Activity Period Day"]
}))

# Sort by date
test_df.sort_values(by="Activity Period Date", inplace=True)

# One hot encoding
test_df = pd.get_dummies(test_df, columns=["GEO Summary", "Cargo Type Code", "Cargo Aircraft Type", "Activity Type Code"])

test_df.drop(columns=exclude, inplace=True, axis=1)
```

# Finding the Best Model

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# Split independent and dependent features for train and test set

# X = train_df.drop(['Activity Period', 'Operating Airline', 'Operating Airline IATA Code',
#        'Published Airline', 'Published Airline IATA Code', 'GEO Region',
#        'Activity Type Code', 'Cargo Metric TONS', 'Activity Period Date',
#         'Activity Period Month', 'Activity Period Day', 'Cargo Type Code_Cargo', 'Cargo Type Code_Express',
#         'Cargo Type Code_Mail', 'Cargo Aircraft Type_Combi', 'Cargo Aircraft Type_Freighter',
#         'Cargo Aircraft Type_Passenger', 'GEO Summary_International', 'GEO Summary_Domestic'], axis=1)

X = train_df.drop(["Cargo Metric TONS", "Activity Period", "Activity Period Date"], axis=1)
train_df.drop(["Activity Period", "Activity Period Date"], axis=1, inplace=True)

# exclude = [col for col in train_df.columns if train_df[col].dtype != "int64"]
# print(exclude)
# print(train_df["Cargo Metric TONS"].dtype)
# X = train_df.drop(exclude, axis=1)

# X = train_df.drop(['Activity Period', 'Operating Airline', 'Operating Airline IATA Code',
#        'Published Airline', 'Published Airline IATA Code', 'GEO Region',
#        'Activity Type Code', 'Cargo Metric TONS', 'Activity Period Date', 'Activity Period Year',
#         'Activity Period Month', 'Activity Period Day', 'Cargo Type Code_Cargo', 'Cargo Type Code_Express',
#         'Cargo Type Code_Mail', 'Cargo Aircraft Type_Combi', 'Cargo Aircraft Type_Freighter',
#         'Cargo Aircraft Type_Passenger'], axis=1)

y = train_df["Cargo Metric TONS"]
```

```python
# Split train and test sets

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=101)
```

```python
X_train
```

```python
y_train
```

```python
# Fit with train data
lr = LinearRegression()
lr.fit(X_train, y_train)
predictions = lr.predict(X_test)
lr.score(X_test, y_test)
```

```python
# Evaluation
print('mean_squared_error : ', mean_squared_error(y_test, predictions))
print('mean_absolute_error : ', mean_absolute_error(y_test, predictions))
```

```python
# Decision Tree
from sklearn import tree

dtr = tree.DecisionTreeRegressor()
dtr.fit(X_train, y_train)
dtr_pred = dtr.predict(X_test)
dtr.score(X_test, y_test)
```

```python
# Evaluation
print('mean_squared_error : ', mean_squared_error(y_test, dtr_pred))
print('mean_absolute_error : ', mean_absolute_error(y_test, dtr_pred))

extracted_MSEs = dtr.tree_.impurity # The Hidden magic is HERE

for idx, MSE in enumerate(dtr.tree_.impurity):
    print("Node {} has MSE {}".format(idx,MSE))
```

```python
# Random Forest
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor(random_state=0)
rfr.fit(X_train, y_train)
rfr_pred = rfr.predict(X_test)
rfr.score(X_test, y_test)
```

```python
# Evaluation
print('mean_squared_error : ', mean_squared_error(y_test, rfr_pred))
print('mean_absolute_error : ', mean_absolute_error(y_test, rfr_pred))
```

```python
# Knn Regressor
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor(n_neighbors=15)
knn.fit(X_train, y_train)
knn_pred = knn.predict(X_test)
knn.score(X_test, y_test)
```

```python
# Evaluation
print('mean_squared_error : ', mean_squared_error(y_test, knn_pred))
print('mean_absolute_error : ', mean_absolute_error(y_test, knn_pred))
```

```python
from sklearn.ensemble import GradientBoostingRegressor

gbr = GradientBoostingRegressor(random_state=0)
gbr.fit(X_train, y_train)
gbr.score(X_test, y_test)
```
