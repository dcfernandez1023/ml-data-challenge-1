```python
from sklearn.preprocessing import LabelBinarizer

lb = LabelBinarizer()
lb.fit(df['ocean_proximity'])
op_array = lb.transform(df['ocean_proximity'])
op_array
```

```python
import numpy as np

lb.inverse_transform(np.array([[0, 1, 0, 0, 0]]))
```

```python
df1H = pd.concat([df, pd.DataFrame(op_array)], axis=1)
df1H
```

```python
df = df1H.drop(['ocean_proximity'], axis=1)
```

```python
df
```

```python
def train_test_split(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_size]
    train_indices = shuffled_indices[test_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
train_set, test_set = train_test_split(df, 0.2)
```

```python
test_set
```

```python
train_y = train_set.median_house_value
test_y = test_set.median_house_value
```

```python
train_set.drop(columns=['median_house_value'], axis=1, inplace=True)
test_set.drop(columns=['median_house_value'], axis=1, inplace=True)
```

```python
test_set
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(train_set, train_y)
```

```python
hyp = reg.predict(test_set)
hyp
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(test_y, hyp)
```

```python
mse
```

```python
np.sqrt(mse)
```

```python
reg.score(train_set, train_y)
```
