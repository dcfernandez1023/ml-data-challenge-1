```python
mean_df = pd.DataFrame(train_df)
mean_df["Activity Period Year"] = mean_df["Activity Period"].astype(str).str.slice(0, 4).astype(int)
mean_df["Activity Period Month"] = mean_df["Activity Period"].astype(str).str.slice(4, 6).astype(int)
mean_df["Activity Period Day"] = 1

mean_df["date"] = pd.to_datetime(pd.DataFrame({
    "year": mean_df["Activity Period Year"],
    "month": mean_df["Activity Period Month"],
    "day": mean_df["Activity Period Day"]
}))
mean_df.drop(columns=["Activity Period Year", "Activity Period Month", "Activity Period Day", "Activity Period"], axis=1, inplace=True)
mean_df
```

```python
international_means = mean_df.where(mean_df["GEO Summary"] == "International").groupby("date")["Cargo Metric TONS"].agg("mean")
```

```python
domestic_means = mean_df.where(mean_df["GEO Summary"] == "Domestic").groupby("date")["Cargo Metric TONS"].agg("mean")
```

```python
domestic_means
```

```python
domestic_avgs = []
mean_df.sort_values(by="date", inplace=True)

for r_index, row in mean_df.iterrows():
    date = row['date']
    for index, val in domestic_means.items():
        if date == index:
            domestic_avgs.append(val)

mean_df["domestic_avg"] = domestic_avgs
mean_df
```

```python
international_avgs = []
mean_df.sort_values(by="date", inplace=True)

for r_index, row in mean_df.iterrows():
    date = row['date']
    for index, val in international_means.items():
        if date == index:
            international_avgs.append(val)

mean_df["international_avg"] = international_avgs

mean_df
```

```python
import numpy as np

mean_df["geo_summary_avg"] = np.where(mean_df["GEO Summary"] == "Domestic", mean_df["domestic_avg"], None)
mean_df["geo_summary_avg"] = np.where(mean_df["GEO Summary"] == "International", mean_df["international_avg"], mean_df["domestic_avg"])
# mean_df.drop(columns=["domestic_avg", "international_avg"], axis=1, inplace=True)
```
